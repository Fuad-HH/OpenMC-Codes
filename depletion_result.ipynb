```python
#Reading the depletion_results file
import openmc
import openmc.deplete
import matplotlib.pyplot as plt
results = openmc.deplete.ResultsList.from_hdf5("./depletion_results.h5")
```

```python
#keff variation with time and keff values
time, k = results.get_eigenvalue()
time /= (24 * 60 * 60)
k
```

```python
# plotting keff vs time
plt.errorbar(time, k[:, 0], yerr=k[:, 1])
plt.xlabel("Time [d]")
plt.ylabel("$k_{eff}\pm \sigma$");
```

```python
#reading data for various atom concentration, this is total atoms (concentration*volume)
_time, u23 = results.get_atoms("4", "U233")
_time, u25 = results.get_atoms("4", "U235")
_time, u28 = results.get_atoms("4", "U238")
_time, pu29 = results.get_atoms("4", "Pu239")
_time, pu20 = results.get_atoms("4", "Pu240")
_time, pu21 = results.get_atoms("4", "Pu241")
_time, pu22 = results.get_atoms("4", "Pu242")
_time, xe15 = results.get_atoms("4", "Xe135")
_time, sm19 = results.get_atoms("4", "Sm149")
```
